```python
from langchain_community.document_loaders import PyPDFLoader
from dotenv import load_dotenv

load_dotenv('../../.env')

loader = PyPDFLoader('./docs/produktspesifikasjoner_ar50_v3_produktspesifikasjon-nibio-ar50-20220523_.pdf')
pages = loader.load_and_split()
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

faiss_index = FAISS.from_documents(pages, OpenAIEmbeddings())
docs = faiss_index.similarity_search("Verdier for vegetasjonsdekke", k=2)
for doc in docs:
    print(str(doc.metadata["page"]) + ":\n", doc.page_content)
```

```python
from langchain.text_splitter import CharacterTextSplitter
import os

pdf_folder_path = "./docs/"
documents = []ø
for file in os.listdir(pdf_folder_path):
    if file.endswith('.pdf'):
        pdf_path = os.path.join(pdf_folder_path, file)
        loader = PyPDFLoader(pdf_path)
        documents.extend(loader.load())
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=10)
chunked_documents = text_splitter.split_documents(documents)
```

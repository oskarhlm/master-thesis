```python
from shapely.geometry import shape
import geopandas as gpd

# Load Innlandet boundary and churches
innlandet_boundary_gdf = gpd.read_file(
    '/tmp/tmpl0e8z051/innlandet_boundary.geojson')
churches_points_gdf = gpd.read_file(
    '/tmp/tmpl0e8z051/churches_innlandet_points.geojson')

# Assuming Innlandet boundary is a single polygon, use the first feature for intersection
innlandet_polygon = innlandet_boundary_gdf.geometry.iloc[1]

# Find intersection of churches within Innlandet
intersection = gpd.overlay(churches_points_gdf, gpd.GeoDataFrame(
    geometry=[innlandet_polygon]), how='intersection')

# Save the result to a new GeoJSON
intersection_file = '/tmp/tmpl0e8z051/churches_innlandet_points_intersection.geojson'
intersection.to_file(intersection_file, driver='GeoJSON')

# len(intersection_file)
intersection
```
